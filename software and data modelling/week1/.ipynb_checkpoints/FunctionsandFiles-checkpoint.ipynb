```python
# Function addInterest(balances, rate)
# This function updates a list of dollar amounts by calculating and adding an interest amount to each amount in the list
# Input: balances (list of dollar amounts), rate (interest rate used in calculation)
# Output: None (balances list is updated due to call-by-reference)
def addInterest(balances, rate):
    for i in range(len(balances)):
        balances[i] = balances[i] * (1+rate)

# Function getinput1()
# This function prompts the user to enter an interest rate and a collection of dollar amounts
# As user enters dollar amounts they are appended onto a list that will be used by the addInterest function
# Input: None
# Output: None
def getinput1():
    rate = eval(input("Enter interest rate:"))
    amounts = []
    amount = eval(input("Enter positive dollar amounts (Enter 0 to quit): "))
    while amount > 0:
        amounts.append(amount)
        amount = eval(input("Enter positive dollar amounts (Enter 0 to quit): "))
    print(amounts)
    print(rate)
    
#getinput1()
#addInterest(amounts, rate)

# Function getinput2()
# This function prompts the user to enter an interest rate and a collection of dollar amounts
# Dollar amounts  are appended onto a list that will be used by the addInterest function
# Input: None
# Output: rate (interest rate entered by the user), amounts (list of dollar values entered by the user)
def getinput2():
    rate = eval(input("Enter interest rate:"))
    amounts = []
    amount = eval(input("Enter positive dollar amounts (Enter 0 to quit): "))
    while amount > 0:
        amounts.append(amount)
        amount = eval(input("Enter positive dollar amounts (Enter 0 to quit): "))
    return rate, amounts
    
#rate, amounts = getinput2()
#addInterest(amounts, rate)
#print(amounts)

# Function getinput3()
# This function prompts the user to enter an interest rate and a collection of dollar amounts
# Dollar amounts are appended onto a list that will be used by the addInterest function
# Error checking code has been added to validate user input
# Input: None
# Output: rate (interest rate entered by the user), amounts (list of dollar values entered by the user)
def getinput3():
    rate = eval(input("Enter interest rate:"))
    while rate<=0:
        rate = eval(input("Invalid interest rate.  Enter valid interest rate:"))        
    amounts = []
    amount = eval(input("Enter positive dollar amounts (Enter 0 to quit): "))
    while amount > 0:
        amounts.append(amount)
        amount = eval(input("Enter positive dollar amounts (Enter 0 to quit): "))
    return rate, amounts
    
#rate, amounts = getinput3()
#addInterest(amounts, rate)
#print(amounts)

# Function getinput4()
# This function prompts the user to enter an interest rate and a collection of dollar amounts
# Dollar amounts are appended onto a list that will be used by the addInterest function
# Exception handling code has been added to validate user input
# Input: None
# Output: rate (interest rate entered by the user), amounts (list of dollar values entered by the user)
def getinput4():
    rate = -1
    while rate < 0:
        try:
            rate = eval(input("Enter interest rate:"))
        except:
            print("Invalid interest rate.  Try again.")
        if rate < 0:
            print("Invalid interest rate.  Try again.")
    amounts = []
    amount = eval(input("Enter positive dollar amounts (Enter 0 to quit): "))
    while amount > 0:
        amounts.append(amount)
        amount = eval(input("Enter positive dollar amounts (Enter 0 to quit): "))
    return rate, amounts
    
#rate, amounts = getinput4()
#addInterest(amounts, rate)
#print(amounts)
```

#  Files

o There are two types of files in Python, binary and text files.

o There are 4 types of operations that can be handled by Python on files:

    o Open - Used to assign a file handle (pointer) to the file being opened
    
           - Must specify name of file and mode.  There are multiple modes:
           
                'r' – Read Mode: Used only to read data from a file.
                
                'w' – Write Mode: Used to write data into a file.  Write mode overwrites the data present in the file.
                
                'a' – Append Mode: Used to append data to the file. Data will be appended at the end of the file pointer.
                
               'r+' – Read or Write Mode: Used to write or read the data from the same file.
               
               'a+' – Append or Read Mode: Used to read data from the file or append the data into the same file.

    o Read - there are 3 ways to read from a file
    
        read() - reads the entire file at once and stores in a Python string
        
        readline() - reads one line of the file at a time.  Must utilize a loop to read in the entire file
           
        readlines() - reads entire file into a list where each element in the list is a  line in the file

    o Write - there are 3 ways to write to a file
    
        write(string) - writes input string as a single line in the file
        
        writelines(list) - writes each element in input list as a line in the file
        
        print(filehandle, string) - uses print method to write to the file

    o Close - files must be closed before being opened again (files automatically closed on program termination)

    o Python also supports:

        o Rename

        o Delete


```python
ifile = input("Enter input file name  -> ")
infile = open(ifile, "r")
names = infile.read()
nameslist = names.split("\n")
print(nameslist)


ifile = input("Enter input file name  -> ")
ofile = input("Enter output file name ->")

# open files
infile = open(ifile, "r")
outfile = open(ofile, "w")

# process each line of the input file
for line in infile:
    # get the first and last names from line
    first, last = line.split()
    # create the username
    username = (first[0]+last[:7]).lower()
    # write it to the output file
    print(username, file=outfile)

# close files
infile.close()
outfile.close()

print("Usernames have been written to", ofile)
```
