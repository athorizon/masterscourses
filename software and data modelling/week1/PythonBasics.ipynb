# Python Language Basics

o Python uses whitespace (tabs or spaces) to structure code instead of using braces

o Text preceded by the hash mark (pound sign) # is considered comment code and is ignored by the Python interpreter. 

o Indentation used instead of braces

o Colon denotes the start of an indented code block after which all of the code must be identically indented

o Function and object method structure and calls nearly identical to other popular languages

o Everything is an object

o Almost every object in Python has attached methods that have access to the objectâ€™s internal contents

o Primitive data types such as number and strings use call-by-value parameter passing

o Strings are immutable in Python

o Lists and objects are use call-by-reference parameter passing

o Lists

    - Python does not have an array type but lists can be used to mimic arrays
    - Lists are mutable
    - Data items can be of different type
    - The first element in a Python list is the zeroth element.

o if, elif, and else used for selection structures

o for and while are used for repetition structure

```python
my_list = ['p','r','o','b','l','e','m']
my_list.remove('p')

# Output: ['r', 'o', 'b', 'l', 'e', 'm']
print(my_list)

# Output: 'o'
print(my_list.pop(1))

# Output: ['r', 'b', 'l', 'e', 'm']
print(my_list)

# Output: 'm'
print(my_list.pop())

# Output: ['r', 'b', 'l', 'e']
print(my_list)

my_list.clear()

# Output: []
print(my_list)

# Changing and adding Dictionary Elements
my_dict = {'name': 'Jack', 'age': 26}

# update value
my_dict['age'] = 27

#Output: {'age': 27, 'name': 'Jack'}
print(my_dict)

# add item
my_dict['address'] = 'Downtown'

# Output: {'address': 'Downtown', 'age': 27, 'name': 'Jack'}
print(my_dict)

# Removing elements from a dictionary

# create a dictionary
squares = {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# remove a particular item, returns its value
# Output: 16
print(squares.pop(4))

# Output: {1: 1, 2: 4, 3: 9, 5: 25}
print(squares)

# remove an arbitrary item, return (key,value)
# Output: (5, 25)
print(squares.popitem())

# Output: {1: 1, 2: 4, 3: 9}
print(squares)

# remove all items
squares.clear()

# Output: {}
print(squares)

# delete the dictionary itself
del squares
```
