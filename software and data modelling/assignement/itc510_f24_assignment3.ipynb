```python
import pandas as pd
import numpy as np

# exam_data

def genDataFrame(dic,narr):

    # print(type(narr))

    if(type(dic)!=dict):
        return "The first parameter should be of type dict"
    if(type(narr)!=np.ndarray):
        return "The second parameter should be of type list to be used as labels"

    # errMsg
    lengths = [len(val) for val in dic.values()]
    lengths.append(len(narr))
    lengths = set(lengths)

    types = [type(val) for val in dic.values()]
    types = set(types)
    

    # print(types)
    if(len(types) != 1 or types.pop() != list):
        return "The data in dictonary is not of valid types"
    if(len(lengths) == 1):
        df = pd.DataFrame(data = dic, index=narr)
        return df
    return "There is an error in the lengths of the data"

# print(genDataFrame(exam_data,labels))
```
